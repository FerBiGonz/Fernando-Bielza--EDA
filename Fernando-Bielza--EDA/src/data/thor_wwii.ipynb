```python
#Aquí creo la variable atacantes, para ver cuantos ataques hacen los paises
df['COUNTRY_FLYING_MISSION'] = df['COUNTRY_FLYING_MISSION'].fillna('Unknown')#<---Esto lo usamos para sustituir los NAN
atacantes=df["COUNTRY_FLYING_MISSION"]
atacantes = atacantes.value_counts()
atacantes
```

```python
#Aquí estoy extrayendo el dato del país más bombardeador, y saber su procentaje en el compluto global

total_bombardeos=df.shape[0]
bombas_USA=atacantes[0]

print ("Bombas detonadas por USA:", bombas_USA, "-", "Total de bombas detonadas:", total_bombardeos)

porcentaje_bombas_USA=(100/total_bombardeos)*bombas_USA
print ("USA detonó el %s del total de bombas detonadas por los Aliados" % porcentaje_bombas_USA)
```

```python
for i in atacantes:
    porcentaje_bombas=(100/df.shape[0])*i
    print ("Detonaciones totales:", i)
    print ("Porcentaje de las detonaciones totales:", porcentaje_bombas)
    
```

```python
#Suma total de los bombardeos dentro de la lista de autores ¡SORPRESA! 

suma=0
for i in atacantes:
    suma=suma+i
print (suma)

porcentaje_bombas_2=(100/total_bombardeos)*(total_bombardeos-suma)

print ("Hubo", total_bombardeos, "detonaciones totales, pero solo", suma,
       "son reconocidas, lo que implica", (total_bombardeos-suma),
       "sin autor, que es un", porcentaje_bombas_2, "% del total")
```
